```python
import os, sys

# Define the path to your project's root and the src directory
project_root = "/content/drive/MyDrive/flood-detect"
src_path = os.path.join(project_root, "src")

# Add src folder to Python path - keeping this as a safety
if src_path not in sys.path:
    sys.path.append(src_path)

# Try direct import and inspect
try:
    import eda
    print("Module 'eda' imported successfully.")
    print("Contents of 'eda' module:", dir(eda))

    # If the module is imported, check for run_eda and call it
    if 'run_eda' in dir(eda):
        print("Function 'run_eda' found in 'eda' module.")
        # Assuming df is available from previous cells
        results = eda.run_eda(df)
        print("run_eda executed successfully.")
        print(results)
    else:
        print("Function 'run_eda' not found in 'eda' module.")

except ModuleNotFoundError:
    print("Error: Module 'eda' not found. Please ensure 'eda.py' is in the specified src directory and accessible.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
